```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
train_data = pd.read_csv("../data/processed_train.csv")
test_data = pd.read_csv("../data/processed_test.csv")
```

```python
data = pd.concat([train_data, test_data], ignore_index=True)
```

```python
plt.figure(figsize=(20, 6))
sns.countplot(data=data, x='category', order=data['category'].value_counts().index, palette='viridis')
plt.title("Category Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(x=data['sub_category'].value_counts().index, y=data['sub_category'].value_counts().values, palette="viridis")
plt.title("Sub-Category Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=data, y='sub_category', hue='category', dodge=False)
plt.title("Category-Sub-Category Distribution")
plt.legend(title='Category')
plt.show()
```

```python
text = " ".join(data['crimeaditionalinfo'].astype(str))

plt.figure(figsize=(10, 8))
wordcloud = WordCloud(width=800, height=400, background_color="white").generate(text)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title("Top Words in Crime Information")
plt.show()
```

```python
category_summary = data.groupby('category').size().reset_index(name='Count').sort_values('Count', ascending=False)
subcategory_summary = data.groupby('sub_category').size().reset_index(name='Count').sort_values('Count', ascending=False)
```

```python
print("Category Summary Table")
print(category_summary)
```

```python
print("\nSub-Category Summary Table")
print(subcategory_summary)
```
